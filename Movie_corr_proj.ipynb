```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv('/Users/rushialmalhotra/Downloads/movies.csv')

df.head()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df.dtypes
```

```python
df = df.dropna()
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
df['year_correct'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df
```

```python
df = df.sort_values(by=['gross'] , inplace =False , ascending = False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
df.head()
```

```python
plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross')

plt.xlabel('Budget for films')

plt.ylabel('Gross earnings')

plt.show()
```

```python
sns.regplot(x='budget',y='gross', data = df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
df.corr(method = 'pearson')
```

```python
correlational_matrix = df.corr(method = 'pearson')
```

```python
sns.heatmap(correlational_matrix, annot = True)

plt.title('Correlation Matric for Numerical Features ')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
```

```python
df_numerized.head(10)
```

```python
correlational_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(correlational_matrix, annot = True)

plt.title('Correlation Matric for Numerical Features ')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs

sorted_pairs = corr_pairs.sort_values()
sorted_pairs

high_corr = sorted_pairs[sorted_pairs > 0.5]
high_corr
```

```python
high_corr = high_corr[high_corr != 1.0]
high_corr
```
